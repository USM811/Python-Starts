# 2 - Python Data Types

### Python automatically decide the data-type of a variable, when we assign any value to the variable

 ### str data-type

```python
x = "Hello World"

#display x:
print(x)

#display the data type of x:
print(type(x))
```

### setting the data-type ' str ' specifically

```python
a = str('hello world')

print(x)

print(type(x))
```

### int data-type

```python
x = 20

print(x)

print(type(x))
```

### setting the data-type ' int ' specifically

```python
x = int(20)

print(x)

print(type(x))
```

### float data-type

```python
x = 20.5

print(x)

print(type(x))
```

### setting the data-type 'float' specifically

```python
x = float(20)

print(x)

print(type(x))
```

### complex data-type

```python
x = -2 + 3j

print(x)

print(type(x))
```

### setting the data-type 'complex' specifically

```python
# Creating a complex number, with '0' imaginary part.

x = complex(4)

print(x)

print(type(x))
```

### list data-type

```python
x = ['apple', 'banana', 'cherry']

print(x)

print(type(x))
```

### setting the data-type 'list' specifically

```python
x = list(('mango', 'grapes', 'peach'))

print(x)

print(type(x))
```

### tuple data-type

```python
x = ('apple', 'mango', 'water-melon')

print(x)

print(type(x))
```

### setting the data-type 'tuple' specifically

```python
x = tuple(('melon', 'grapes', 'apple'))

print(x)

print(type(x))
```

### range data-type

```python
x = range(6)

print(x)

print(type(x))
```

```python
x = range(-6, 6)

print(x)

print(type(x))
```

### dict data-type

```python
x = {'name': 'Jhon', 'age': 36}

print(x)

print(type(x))
```

### setting the data-type 'dict' specifically - use single paranthesis

```python
x = dict(name='Usman', age=35)

print(x)

print(type(x))
```

### ' set ' data-type

```python
x = {'apple', 'banana', 'orange'}

print(x)

print(type(x))
```

### setting the data-type 'set' specifically

```python
x = set(('mango', 'melon', 'grapes'))

print(x)

print(type(x))
```

### ' frozenset' data-type

```python
x = frozenset({'apple', 'banan', 'orange'})

print(x)

print(type(x))
```

### setting the data-type ' frozenset ' specifically

```python
x = frozenset(('melon', 'mango', 22))

print(x)

print(type(x))
```

###  ' bool ' data-type

```python
x = True

print(x)

print(type(x))
```

### setting the data-type ' bool ' specifically

```python
x = bool(5)

print(x)

print(type(x))

y = bool(-5)
print(y)
print(type(y))
```

```python
x = bool(0)

print(x)

print(type(x))
```
