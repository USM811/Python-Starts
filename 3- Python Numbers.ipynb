# 3 - Python Numbers

```python
x = -10
y = 2.8
z = 2-7j

print(x)
print(type(x), '\n')

print(y)
print(type(y), '\n')

print(z)
print(type(z), '\n')
```

### Float can also be scientific numbers with an "e" to indicate the power of 10.

```python
a = 35e3
b = 12E-4
c = -87.7e100

print(a)
print(type(a), '\n')

print(b)
print(type(b), '\n')

print(c)
print(type(c), '\n')
```

### Type Conversion

```python
x = 1        # int
y = 2.8      # float
z = 1j       # complex

#convert from int to float:
a = float(x)

#convert from float to int:
b = int(y)

#convert from int to complex:
c = complex(x)

print('Before Conversion, x = ', x , '\n', type(x), '\n')
print('After Conversion, x = ', a, '\n', type(a), '\n')

print('----------------------------')

print('Before Conversion, y = ', y , '\n', type(y), '\n')
print('After Conversion, y = ', b, '\n', type(b), '\n')

print('----------------------------')

print('Before Conversion, x = ', x , '\n', type(x), '\n')
print('After Conversion, x = ', c, '\n', type(c), '\n')

print('----------------------------')
```

### Random Number:

### Python does not have a random() function to make a random number, 

### but Python has a built-in module called 'random' that can be used to make random numbers:

```python
import random

print(random.randrange(-10, 10))
```
