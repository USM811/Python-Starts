### Exted List - extend()

#### extend() method used to extend the list by add another list.

```python
thislist = ["apple", "banana", "cherry"]

newlist = ["orange", 'mango']

thislist.extend(newlist)

print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"]

thislist.extend(["orange", 'mango'])

print(thislist)
```

#### There will be problem, if we try to add single element(string) in the list by using extend() method.

```python
thislist = ['apple', 'banana', 'cherry']

thislist.extend('mango')

print(thislist)
```

#### we can resolve this issue, by using the list.

```python
thislist = ['apple', 'banana', 'cherry']

thislist.extend(['mango'])

print(thislist)
```

## List Comprehension:

#### List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

#### Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.

#### Without list comprehension you will have to write a for statement with a conditional test inside:

```python
fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango', 'grapes']

newlist =[]

for x in fruits:
    if 'a' in x:
        newlist.append(x)

print(newlist)
```

### With list comprehension you can do all that with only one line of code:

```python
fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango', 'grapes']

newlist = [x for x in fruits if 'a' in x]

print(newlist)
```

```python
fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango', 'grapes']

print(fruits, '\n')

newlist = [x for x in fruits if x != 'apple']

print(newlist)
```

#### Condition is Optional

```python
fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango', 'grapes']

print(fruits, '\n')

newlist = [x for x in fruits]

print(newlist)
```

### List Comprehension can be done through any iterable (i.e. list, tuple, set)

#### using the range() function to create an iterable:

```python
newlist = [x for x in range(10)]

print(newlist)
```

#### Same example, but with a condition:

```python
newlist = [x for x in range(50) if x % 5 == 0]

print(newlist)
```

#### The expression is the current item in the iteration, but it is also the outcome, 

#### which you can manipulate before it ends up like a list item in the new list:

```python
fruits = ['apple', 'banana', 'cherry', 'melon', 'kiwi', 'grapes']

print(fruits, '\n')

newlist = [x.upper() for x in fruits]

print(newlist)
```

#### You can set the outcome to whatever you like:

```python
fruits = ['apple', 'banana', 'cherry', 'melon', 'kiwi', 'grapes']

newlist = ['hello' for x in fruits]

print(newlist)
```

#### The expression can also contain conditions, not like a filter, but as a way to manipulate the outcome:

#### Return "orange" instead of "banana":

```python
fruits = ['apple', 'banana', 'cherry', 'melon', 'kiwi', 'grapes']

newlist = [x if x != 'banana' else 'orange' for x in fruits]

print(newlist)
```

#### The expression in the example above says:

#### "Return the item if it is not banana, if it is banana return orange".

## Python - List Sorts:

### Sort the List Alphanumarically - sort()

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]

print('Before Sorting the List is => ', thislist, '\n')

thislist.sort()

print('After Sorting the List is => ', thislist)
```

### Sort the List Numarically - sort()

```python
thislist = [100, 50, 65, 82, 23]

print('Before Sorting the List is => ', thislist, '\n')

thislist.sort()

print('After Sorting the List is => ', thislist)
```

### Sort the List Descending - sort(reverse = True)

```python
thislist = [100, 50, 65, 82, 23]

print('Before Sorting the List is => ', thislist, '\n')

thislist.sort(reverse = True)

print('After Sorting the List is => ', thislist)
```

### Another Way - apply reverse() method on the sorted list.

```python
thislist = [100, 50, 65, 82, 23]

print('Before Sorting the List is => ', thislist, '\n')

thislist.sort()

print('After Sorting the List is => ', thislist, '\n')

thislist.reverse()

print('After apply the reverse() method on the sorted list => ', thislist)
```

### Customize Sort Function - sort(key = function_name)

### Sort the list based on how close the every number is to 50:

```python
def my_fun(n):
    return abs(n - 50)

list_1 = [100, 45, 65, 23, 82, 99]

print('Before Sorting, the list is =>', list_1, '\n')

list_1.sort(key = my_fun)

print('After Sorting, the list is =>', list_1)
```

### Short Method of Customize Sort Function

```python
this_list = [99, 83, 23, 66, 44, 100]

print('Before Sorting, the list is =>', this_list, '\n')

this_list.sort(key = lambda x: abs(x-50))

print('After Sorting, the list is =>', this_list)
```

### [ Case Sensitive Sorting ] Vs [ Case Insensitive Sorting ]

#### By default the sort() method is case sensitive, 

#### firtly it sort the capital letters, and then it sort the small letter:

```python
thislist = ['banana', 'apple', 'Cherry', 'Kiwi']

print('Before Sorting, the list is =>', thislist, '\n')

thislist.sort()

print('After Sorting, the list is =>', thislist, '\n')
```

#### So if you want a case-insensitive sort function,

#### use str.lower as a key function:

```python
thislist = ['banana', 'apple', 'Cherry', 'Kiwi']

print('Before Sorting, the list is =>', thislist, '\n')

thislist.sort(key = str.lower)

print('After Sorting, the list is =>', thislist, '\n')
```

### Reverse Order of the List:

```python
thislist = ['banana', 'apple', 'Cherry', 'Kiwi']

print('Before apply reverse() function, the list is =>', thislist, '\n')

thislist.reverse()

print('After apply reverse() function, the list is =>', thislist, '\n')
```

### Copy a List:

### You cannot copy a list simply by typing list_1 = list_2, because: the memory address of ( list_1 ) will be

### refer to ( list_2 ), and changes made in ( list_2 ) will automatically also be made in ( list_1 ).

### There are  two ways to copy a List without referece the memory address.

### 1- using copy() method

### 2- using list() method

```python
thislist = ["apple", "banana", "cherry"]

mylist = thislist.copy()

mylist.append('water-melon')

print('\nThe Copied and changed List is ', mylist, '\n')

print('The Original List is Not-affected', thislist)
```

```python
thislist = ["apple", "banana", "cherry"]

mylist = list(thislist)

mylist.append('grapes')

print('\nThe Copied and changed List is ', mylist, '\n')

print('The Original List is Not-affected', thislist)
```
