# 4 - Python Casting

### Casting in python is therefore done using constructor functions:

### int() - constructs an integer number from an integer literal, a float literal (by rounding down to the previous 

### whole number), or a string literal (providing the string represents a whole number)

```python
a = 1
b = 2.8
c = "3"

# Type Casting
x = int(a)
y = int(b)
z = int(c)

print('Before Casting, a = ', a, '\n', type(a), '\n')
print('After Casting, a = ', x, '\n', type(x), '\n')
print('--------------------------')
print('Before Casting, b = ', b, '\n', type(b), '\n')
print('After Casting, b = ', y, '\n', type(y), '\n')
print('--------------------------')
print('Before Casting, c = ', c, '\n', type(c), '\n')
print('After Casting, c = ', z, '\n', type(z), '\n')
print('--------------------------')
```

### float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string 

### represents a float or an integer)

```python
x = 1
y = 2.8
z = "3"
w = "4.2"

cx = float(x)
cy = float(y)
cz = float(z)
cw = float(w)

print('Before Casting, x = ', x, '\n', type(x), '\n')
print('After Casting, x = ', cx, '\n', type(cx), '\n')
print('--------------------------')
print('Before Casting, y = ', y, '\n', type(y), '\n')
print('After Casting, y = ', cy, '\n', type(cy), '\n')
print('--------------------------')
print('Before Casting, z = ', z, '\n', type(z), '\n')
print('After Casting, z = ', cz, '\n', type(cz), '\n')
print('--------------------------')
print('Before Casting, w = ', w, '\n', type(w), '\n')
print('After Casting, w = ', cw, '\n', type(cw), '\n')
print('--------------------------')
```

### str() - constructs a string from a wide variety of data types, including strings, integer literals and float 

### literals.

```python
x = 'string-1'
y = 2
z = 3.0

# Type Casting
cx = str(x)
cy = str(y)
cz = str(z)

print('Before Casting, x = ', x, '\n', type(x), '\n')
print('After Casting, x = ', cx, '\n', type(cx), '\n')
print('--------------------------')
print('Before Casting, y = ', y, '\n', type(y), '\n')
print('After Casting, y = ', cy, '\n', type(cy), '\n')
print('--------------------------')
print('Before Casting, z = ', z, '\n', type(z), '\n')
print('After Casting, z = ', cz, '\n', type(cz), '\n')
print('--------------------------')
```
