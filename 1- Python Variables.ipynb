### The concept of 'global variable': create a variable outside the function and use it inside the function.

```python
a = 'awesome'

def myfunction():
    print('Python is' , a)

# call the function
myfunction()
```

### If you create a variable with the same name inside a function, this variable will be local, and can only be used inside the function. 

### The global variable with the same name will remain as it was, global and with the original value.

```python
a = 'awesome'

def myfunction():
    a = 'fantastic'
    print('Python is', a)

# Call the function
myfunction()

# Use of 'global variable' outside the functin.
print('Python is', a)
```

### Use of 'global' keyword:

### Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.

### To create a global variable inside a function, you can use the "global" keyword.

```python
def myfunction():
    global x
    x = 'fantastic'
    print('Python is', x , ', from inside the function.')

# Calling the function
myfunction()

# Using the variable 'x' which defined inside the function with keyword 'global'
print('Python is',x , ', accessing the function-variable outside the function.')
```

### 'global' keyword also used to change the value of (global-variable) inside the function:

```python
x = 'awesome'

def myfunction():
    global x
    x = 'fantastic'
    print('Python is', x)

# function calling
myfunction()

# Using global variable to confirm the value of global-variable is changed.
print('Python is', x)
```
