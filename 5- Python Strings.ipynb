### String Methods:

### Python has a set of built-in methods that you can use on strings.

### The strip() method removes any whitespace from the beginning or the end:

```python
a = '        hello my dear friend!        '

print(a)
print(a.strip())
```

### The lower() method returns the string in lower case:

```python
b = 'heLLo mY DeAR FriEnD!'

print(b)
print(b.lower())
```

### The upper() method returns the string in upper case:

```python
print(b.upper())
```

### The capitalize() method returns the string with only first character in upper case and remaining characters 

### in lower case:

```python
print(b.capitalize())
```

### The replace() method replaces a string with another string:

```python
print(b)
c = b.replace('heLLo', 'hI')
print(c)
```

### The split() method splits the string into substrings if it finds instances of the separator:

```python
d = c.split(' ')
print(d)
```

### Check String

### To check if a certain phrase or character is present in a string, we can use the keywords ' in ' or ' not in '.

```python
txt = "The rain in Spain stays mainly in the plain"
x = "ain" in txt
print(x)
```

```python
txt = "The rain in Spain stays mainly in the plain"
x = "ain" not in txt
print(x)
```

### we can combine strings and numbers by using the format() method!

### The format() method takes the passed arguments, formats them, and places them in the string 

### where the placeholders {} are:

```python
# Use the formate() method to insert the number into string.

age =36
txt = 'My name is Jhon, and I am {}'

print(txt.format(age))
```

### The format() method takes unlimited number of arguments, and are placed into the respective placeholders:

```python
quantity = 3
item_number = 201390
price = 49.95

my_order = 'I want {} pieces of item number {} for {} dollars.'

output = my_order.format(quantity, item_number, price)

print(output)
```

### You can use index numbers {0} to be sure the arguments are placed in the correct placeholders:

```python
quantity = 3
item_number = 201390
price = 49.95

my_order = 'I want pay {2} dollars to buy the {0} pieces of item number {1}'

output = my_order.format(quantity, item_number, price)

print(output)
```

```python
txt = "Hello \bWorld!"
print(txt)
```

### A backslash followed by three integers will result in a octal value: \000

```python
txt = "\110\145\154\154\157"
print(txt)
```

### A backslash followed by an 'x' and a hex number represents a hex value:

```python
txt = "\x48\x65\x6c\x6c\x6f"
print(txt)
```
