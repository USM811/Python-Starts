### Python Arithmetic Opetators

```python
# Addition
x = 5
y = 3

print(x + y)
```

```python
# Multiplation
x = 5
y = 3

print(x * y)
```

```python
# Division
x = 12
y = 3

print(x / y)
```

```python
# Modulus Operator / Remainder Operator
x = 5
y = 2

print(x % y)
```

```python
# Exponential Operator
x = 2
y = 5

print(x ** y) #same as 2*2*2*2*2
```

### Python Assignment Operators

```python
# Simply Assignment Operator

x = 5

print(x)
```

```python
# Short-hand Form of Increment Operator

x = 5
x += 2     # (x = x + 2)

print(x)
```

```python
# Short-hand Form of Decrement Operator

x = 5
x -= 2     # (x = x - 2)

print(x)
```

```python
# Short-hand Form of Multiple Operator

x =5
x *= 3    # (x = x * 3)

print(x)
```

```python
# Short-hand Form of Division Operator

x = 5
x /= 2    # (x = x / 2)

print(x)
```

```python
# Short-hand Form of "Floor Division Operator"

x = 5
x //=2    # (x = x // 2)

print(x)
```

```python
# Short-hand Form of "Remainder Operator"

x = 5
x %= 2

print(x)
```

```python
# Short-hand Form of "Power Operator"

x = 5
x **=2    # (x = x ** 2)

print(x)
```

```python
# 'AND' Operator (Answer will be 'True(1)', only if both inputs are 'True(1)', otherwise 'False(0)')

x = 5
x &= 3   # (x = x & 3)

print(x)
print(bool(x))
```

```python
# 'OR' Operator (Answer will be 'True(1)', if any one of both inputs are 'True(1)', otherwise 'False(0)')

x = 5
x |= 3   # (x = x | 3)

print(x)
print(bool(x))
```

```python
# Exclusive-OR 'XOR' Operator (Answer will be 'True(1)', only if one of both inputs are 'True(1)')

print('1) ', True ^ False)

print('2) ', False ^ True)

print('3) ', True ^ True)

print('4) ', False ^ False)
```

#### Exclusive-OR

```python
# But when we apply (Exclusive-OR) 'XOR' operator to integers, it behave differently.

# The value will increment by one, by applying 'XOR' operator with 'True' value.

# The value will not increment by one (remain same), by applying 'XOR' operator with 'False' value.

x = 5
print('Before apply \'XOR\' Oerator, x = ', x, '\n')

x ^= 3   # (x = x ^ 3)

print('After apply \'XOR\' Operator with \'True\' value, x = ', x)

print('-----------------------------\n')

y = 10
print('Before apply \'XOR\' Oerator, y = ', y, '\n')

y = y ^ False
print('After apply \'XOR\' Operator with \'False\' value, y = ', y)

print('------------------------------\n')

a = 10
print('Before apply \'XOR\' Oerator, a = ', a, '\n')

a = a ^ True
print('After apply \'XOR\' Operator with \'True\' value, a = ', a)

print('------------------------------\n')

print('There will no change by applying \'XOR\' operator to negative integers or zero.\n')
print('------------------------------\n')

z = -6
print('Before apply \'XOR\' Oerator, z = ', z, '\n')

y = y ^ True
print('After apply \'XOR\' Operator with \'True\' value, z = ', z)

print('------------------------------\n')

b = -1
print('Before apply \'XOR\' Oerator, b = ', b, '\n')

b = b ^ False
print('After apply \'XOR\' Operator with \'False\' value, b = ', b)

print('------------------------------')
```

#### Left Shif Operator

```python
# 'Left Shift' operator convert the given number into binary formate, 
# and drop n-digits from the left side and simply add the n-digits on the right side

x = 5
print('After apply the \'Left-Shift\' Operator, x = ', x, '\n')

print('Before applying the \'Left-Shift\' Operator, the binary form of x is => ', bin(x), '\n')

x <<= 5
print('------------------------------------------------------------------\n')

print('After apply the \'Left-Shift\' Operator, x = ', x, '\n')
print('After apply the \'Left-Shift\' Operator, the binary form of x is => ', bin(x))
```

#### Right-Shift Operator

```python
# 'Right Shift' operator convert the given number into binary formate, 
# and drop n-digits from the right side and simply add the copy of n-digits from the left side, on the left side.

x = 5
print('After apply the \'Left-Shift\' Operator, x = ', x, '\n')

print('Before applying the \'Left-Shift\' Operator, the binary form of x is => ', bin(x), '\n')

x >>= 5
print('------------------------------------------------------------------\n')

print('After apply the \'Left-Shift\' Operator, x = ', x, '\n')
print('After apply the \'Left-Shift\' Operator, the binary form of x is => ', bin(x))
```

### Python Comparison Operators

```python
x = 5
y = 3

print('x = ', x, '\n')
print('y = ', y, '\n')

print("x == y, yeilds: ", x == y, '\n')

print('x != y, yeilds: ', x != y, '\n')

print('x > y, yeilds: ', x > y, '\n')

print('x < y, yeilds: ', x < y, '\n')

print('x >= y, yeilds: ', x >= y, '\n')

print('x <= y, yeilds: ', x <= y, '\n')
```

### Python Logical Operator

```python
x = 5
y = 10
print('x<5 and y>10, yeilds: ', x < 5 and y > 10, '\n')

print('x<5 or y>5, yeilds: ', x < 5 or y > 5, '\n')

print('not(x<5 or y>5), yeilds: ', not(x < 5 or y > 5), '\n')
```

### Python Identity Operator

```python
x = 5
y = 3

print('x is y => ', x is y, '\n')

print('x is not y => ', x is not y , '\n')
```

### Python Membership Operator

```python
list_1 = ['banana', 'mango', 'apple']

print(list_1, '\n')

print('\'mango\' in list_1 => ', 'mango' in list_1, '\n')

print('\'mango\' not in list_1 => ', 'mango' not in list_1, '\n')
```
